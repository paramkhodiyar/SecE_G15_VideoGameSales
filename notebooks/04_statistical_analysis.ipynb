```python
import pandas as pd
from scipy.stats import pearsonr
df = pd.read_csv("../data/processed/video_games_cleaned.csv")
```

```python
pearsonr(df["Critic_Score"], df["Global_Sales"])
```

Positive correlation between critic score and sales suggests review quality impacts consumer purchase decisions.

```python
from scipy.stats import ttest_ind

action_sales = df[df["Genre"]=="Action"]["Global_Sales"]
other_sales = df[df["Genre"]!="Action"]["Global_Sales"]

ttest_ind(action_sales, other_sales)
```

Action games generate significantly higher sales than other genres.
