```python
import pandas as pd
df = pd.read_csv("../data/processed/video_games_cleaned.csv")
```

```python
genre_kpi = df.groupby("Genre").agg(
{
"Global_Sales":"sum",
"Critic_Score":"mean",
"User_Score":"mean"
}
).reset_index()

genre_kpi
```

```python
genre_kpi.to_csv("../data/processed/genre_kpi.csv", index=False)
```

```python
platform_kpi = df.groupby("Platform").agg(
{
"Global_Sales":"sum",
"Critic_Score":"mean"
}
).reset_index()

platform_kpi.to_csv("../data/processed/platform_kpi.csv", index=False)
```

```python
year_kpi = df.groupby("Year_of_Release")["Global_Sales"].sum().reset_index()

year_kpi.to_csv("../data/processed/year_kpi.csv", index=False)
```
